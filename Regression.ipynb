1. **What is Simple Linear Regression?**  
Simple Linear Regression is a statistical method used to model the relationship between two quantitative variables: one independent variable (predictor) and one dependent variable (outcome). It fits a straight line to the data to estimate how changes in the independent variable affect the dependent variables.

2. **What are the key assumptions of Simple Linear Regression?**  
The main assumptions are:
- Linearity: The relationship between the variables is linear.
- Homoscedasticity: The variance of the errors is constant across all levels of the independent variable.
- Independence: Observations are independent of each other.
- Normality: The residuals (errors) are normally distributed.

3. **What does the coefficient m represent in the equation Y=mx+c?**  
The coefficient 'm' is the slope of the line. It shows how much the dependent variable (Y) changes for a one-unit increase in the independent variable (X).

4. **What does the intercept c represent in the equation Y=mx+c?**  
The intercept 'c' is the value of Y when X is zero. It's where the line crosses the Y-axis.

6. **What is the purpose of the least squares method in Simple Linear Regression?**  
The least squares method finds the line of best fit by minimizing the sum of the squared differences (residuals) between the observed values and the values predicted by the line.

7. **How is the coefficient of determination (R²) interpreted in Simple Linear Regression?**  
R² measures how much of the variance in the dependent variable can be explained by the independent variable. An R² of 0 means no explanatory power, while 1 means perfect explanation. For example, R² = 0.8 means 80% of the variance is explained by the model.

8. **What is Multiple Linear Regression?**  
Multiple Linear Regression (MLR) is an extension of simple linear regression that uses two or more independent variables to predict a single dependent variable. It helps estimate how several factors together influence the outcome.

9. **What is the main difference between Simple and Multiple Linear Regression?**  
The key difference is the number of independent variables: simple linear regression uses one, while multiple linear regression uses two or more.

10. **What are the key assumptions of Multiple Linear Regression?**  
- Linearity between dependent and independent variables
- Independence of observations
- Homoscedasticity (constant variance of residuals)
- Normality of residuals
- No (or low) multicollinearity among independent variables

**11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?**  
Heteroscedasticity means that the variance of the residuals (errors) isn’t constant across all levels of the independent variables. This can mess up the reliability of the model’s standard errors, making hypothesis tests and confidence intervals less trustworthy.

**12. How can you improve a Multiple Linear Regression model with high multicollinearity?**  
To deal with high multicollinearity, you can:
- Remove or combine correlated variables
- Use dimensionality reduction techniques like PCA
- Try regularization methods like Ridge or Lasso regression

**13. What are some common techniques for transforming categorical variables for use in regression models?**  
The most common techniques are:
- One-hot encoding (creates binary columns for each category)
- Label encoding (assigns a number to each category)
- Using dummy variables (basically a type of one-hot encoding)

**14. What is the role of interaction terms in Multiple Linear Regression?**  
Interaction terms let you model situations where the effect of one independent variable on the dependent variable depends on another independent variable. It helps capture more complex relationships.

**15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?**  
In simple linear regression, the intercept is the predicted value of Y when X is zero. In multiple regression, it’s the predicted value when all Xs are zero, which sometimes doesn’t make practical sense if zero isn’t a realistic value for all variables.

**16. What is the significance of the slope in regression analysis, and how does it affect predictions?**  
The slope tells you how much the dependent variable is expected to change for a one-unit increase in the independent variable, holding other variables constant. It’s key for making predictions and understanding variable relationships.

**17. How does the intercept in a regression model provide context for the relationship between variables?**  
The intercept gives a baseline value for the dependent variable when all independent variables are zero. It helps set the starting point for predictions, but sometimes it’s not meaningful if zero isn’t a possible value.

**18. What are the limitations of using R² as a sole measure of model performance?**  
R² only tells you how much variance is explained by the model, but it doesn’t show if the model is actually good or if it’s overfitting. It also doesn’t tell you if the predictors are significant or if the model is appropriate.

**19. How would you interpret a large standard error for a regression coefficient?**  
A large standard error means there’s a lot of uncertainty about the true value of the coefficient. It could mean the variable isn’t very useful for predicting the outcome, or there’s not enough data.

**20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?**  
You can spot heteroscedasticity if the spread of residuals increases or decreases as the predicted values change (like a funnel shape) in a residual plot. It’s important to fix because it can make your statistical tests invalid.

**21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?**  
This usually means you’ve added too many variables that don’t actually help predict the outcome. Adjusted R² penalizes for unnecessary predictors, so a big drop means some variables aren’t useful.

**22. Why is it important to scale variables in Multiple Linear Regression?**  
Scaling puts all variables on the same scale, which helps the model converge faster and prevents variables with larger scales from dominating the regression.

**23. What is polynomial regression?**  
Polynomial regression is a type of regression where the relationship between the independent variable(s) and the dependent variable is modeled as an nth-degree polynomial.

**24. How does polynomial regression differ from linear regression?**  
Linear regression models a straight-line relationship, while polynomial regression can model curved relationships by including higher-degree terms.

**25. When is polynomial regression used?**  
It’s used when the data shows a non-linear relationship that a straight line can’t capture.

**26. What is the general equation for polynomial regression?**  
For one variable, it’s:  
$$
y = b_0 + b_1x + b_2x^2 + ... + b_nx^n
$$

**27. Can polynomial regression be applied to multiple variables?**  
Yes, you can have polynomial terms for multiple variables, including interactions and higher-degree terms.

**28. What are the limitations of polynomial regression?**  
It can easily overfit, especially with high-degree polynomials, and it can be sensitive to outliers.

**29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?**  
You can use cross-validation, look at metrics like RMSE or MAE, and check for overfitting by comparing training and validation performance.

**30. Why is visualization important in polynomial regression?**  
Visualization helps you see if the model is capturing the trend in the data or if it’s overfitting or underfitting.
